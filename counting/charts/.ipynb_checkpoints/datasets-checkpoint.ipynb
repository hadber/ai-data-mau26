### Read all .csv files for each dataset

```python
import csv

def load_dataset(file_path):
    with open(file_path, newline='') as csvfile:
        data_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
        return {rows[0]:rows[1] for rows in data_reader}

def vectorize_dataset(file_path):
    as_dict = load_dataset(file_path)
    x = []
    y = []
    for row in as_dict:
        humans, images = row, as_dict[row]
        if not humans.isdigit():
            # this is probably the header, so we skip
            continue
        x.append(int(humans))
        y.append(int(images))

    return x, y
```

```python
kitti_x, kitti_y = vectorize_dataset("kitti.csv")
cocopersons_x, cocopersons_y = vectorize_dataset("cocopersons.csv")
crowdhuman_x, crowdhuman_y = vectorize_dataset("crowdhuman.csv")
```

```python
class Dataset():
    x = []
    y = []
    name = ""
    
    def __init__(self, csv_file, name):
        self.x, self.y = vectorize_dataset(csv_file)
        self.name = name

    def add_index(self, i):
        if i not in self.x:
            self.x.append(i)
            self.y.append(0)

    def sort(self):
        tmp_x, tmp_y = self.x.copy(), self.y.copy()
        for i in range(len(self.x)):
            idx = self.x.index(i)
            tmp_x[i] = self.x[idx]
            tmp_y[i] = self.y[idx]
        self.x, self.y = tmp_x, tmp_y
```

```python
kitti = Dataset("kitti.csv", "KITTI")
cocopersons = Dataset("cocopersons.csv", "COCOPersons")
crowdhuman = Dataset("crowdhuman.csv", "CrowdHuman")

datasets = [kitti, cocopersons, crowdhuman]
```

### Make sure each dataset has the same indices, even if that dataset might not have pictures with, for example, 0 humans.

```python
biggest_index = max(kitti_x+cocopersons_x+crowdhuman_x)

for i in range(biggest_index+1):
    for ds in datasets:
        ds.add_index(i)

for ds in datasets:
    ds.sort()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#species = ("Adelie", "Chinstrap", "Gentoo")
#penguin_means = {
#    'Bill Depth': (18.35, 18.43, 14.98),
#    'Bill Length': (38.79, 48.83, 47.50),
#    'Flipper Length': (189.95, 195.82, 217.19),
#}

x = np.arange(len(kitti.x))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')
plt.yscale('log')

for dataset in datasets:
    offset = width * multiplier
    rects = ax.bar(x + offset, dataset.y, width, label=dataset.name)
#    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Images')
ax.set_xlabel('Humans')
ax.set_title('Number of humans per images contained in datasets')
ax.set_xticks(x + width, x)
ax.legend(loc='upper right', ncols=3)
#ax.set_ylim(0, 1000)

plt.show()
```
